```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar um DataFrame de exemplo
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Idade': [25, 30, 35, 40, 45],
        'Vendas': [1000, 1500, 1200, 1800, 2000]}

df = pd.DataFrame(data)

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
barras = plt.bar(df['Nome'], df['Vendas'], color='skyblue')

# Adicionar os valores das vendas sobre as barras
for barra in barras:
    height = barra.get_height()
    plt.annotate(f'{height}',
                 xy=(barra.get_x() + barra.get_width() / 2, height),
                 xytext=(0, 3),  # Deslocamento vertical do texto
                 textcoords='offset points',
                 ha='center', va='bottom')

plt.xlabel('Nome')
plt.ylabel('Vendas')
plt.title('Vendas por Nome')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajustar o layout para evitar cortes
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar um DataFrame de exemplo
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Vendas': [1000, 1500, 1200, 1800, 2000]}

df = pd.DataFrame(data)

# Calcular a média das vendas
media_vendas = df['Vendas'].mean()

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
barras = plt.bar(df['Nome'], df['Vendas'], color='skyblue')

# Adicionar a média como uma linha no gráfico
plt.axhline(y=media_vendas, color='red', linestyle='--', label='Média de Vendas')

plt.xlabel('Nome')
plt.ylabel('Vendas')
plt.title('Vendas por Nome')
plt.legend()
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajustar o layout para evitar cortes
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar um DataFrame de exemplo
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Vendas': [1000, 1500, 1200, 1800, 2000],
        'Lucro': [200, 300, 240, 360, 400]}

df = pd.DataFrame(data)

# Plotar as colunas 'Vendas' e 'Lucro'
plt.figure(figsize=(10, 6))
plt.plot(df['Nome'], df['Vendas'], label='Vendas', marker='o')
plt.plot(df['Nome'], df['Lucro'], label='Lucro', marker='x')
plt.xlabel('Nome')
plt.ylabel('Valor')
plt.title('Vendas e Lucro por Nome')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Criar um DataFrame de exemplo
data = {'Data': pd.date_range(start='2023-01-01', end='2023-01-03', freq='M'),
        'Dados': range(49)}  # 49 horas em 3 dias

df = pd.DataFrame(data)

# Agrupar os dados por hora e calcular a soma
df_agrupado = df.groupby(pd.Grouper(key='Data', freq='M')).sum()

print(df_agrupado)
```

```python
import pandas as pd

# Criar um DataFrame de exemplo
data = {'Data': pd.date_range(start='2023-01-01', end='2023-01-03', freq='H'),
        'Dados': range(49)}  # 49 horas em 3 dias

df = pd.DataFrame(data)

# Adicionar uma coluna de dados calculados (por exemplo, o quadrado dos dados originais)
df['Dados_calculados'] = df['Dados'] ** 2

# Agrupar os dados por hora e calcular a soma dos dados originais e a média dos dados calculados
df_agrupado = df.groupby(pd.Grouper(key='Data', freq='H')).agg({'Dados': 'sum', 'Dados_calculados': 'mean'})

print(df_agrupado)
```
