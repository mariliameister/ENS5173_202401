```python
# criando uma lista em python
list = [0, 'texto', 2.5]
print (list)
#Dessa maneira é possível fazer a adição de duas strings, sendo que list [1] é o segundo indice do vetor list
print ('variável:' + list[1])

#Adição do primeiro indice da variável list (list[0]) - Lembrar que a indexação no python inicia em 0
print (list[0]+ 1)
```

```python
# Criando uma matriz numpy
import numpy as np

a = np.zeros((5,5))
print(a)
print ('indexação numpy')
print (a[0,0])
```

```python
#criando numeros aleatorios e mostrando a dimensão da matriz
b= np.random.rand (5,5)
print(b)
print('Dimensão da matriz')
print (b.shape) #mostra a dimensão da matriz
```

```python
# criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 #valor mínimo
vmáx = 100 #valor máximo

c = vmin+(vmáx-vmin)*b
print(c)
```

```python
# plotar a matriz 
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha
x= np.array(range(0,np.size(c[0,:]))) # converte range para 
print(x)
plt.plot(c[0,:]) # : significa todas as colunas, ele está plotando os primeiros valores de todas colunas. (comparando pontos)
plt. title ('Gráfico com Valores Inteiros no Eixo X')
plt.xticks(x) #Definindo os valores do eixo x
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1) #criou a figura com os aixos (ax são os eixos)
ax[0].bar(x,c[:,0]) #bar significa que vai plotar um gráfico de barras
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# exportar a figura 
import os
rootFolder = os.getcwd()
print(rootFolder) #mostra a pasta das figuras
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
fig.savefig(figFolder +'Figure01.png')
```
